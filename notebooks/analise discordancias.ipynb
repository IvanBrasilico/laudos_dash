```python
def get_ordem_from_texto(texto: str)-> [int, str]:
    separators = '.)-'
    for separator in separators:
        index = texto[:4].find(separator)
        if index >=0:
            break
    if index >= 0:
        offset = index
    else:
        offset = 2
    num = texto[:offset]
    desc = texto[offset + 1:].strip()
    try:
        num = int(num)
    except ValueError:
        num = 0
    return num, desc
```

```python
for index, linha in dfq.iterrows():
    texto = linha['texto']
    num, desc = get_ordem_from_texto(texto)
    # if num != 0:
    #    print('UPDATE LAUDOS.quesitos SET ordempreferencial=%s, texto="%s" WHERE ID=%s;' % (num, desc, linha['id']))
```

```python
print(get_ordem_from_texto('01. Teste'))
print(get_ordem_from_texto('2) Teste'))
print(get_ordem_from_texto('3- Teste'))
print(get_ordem_from_texto(' Teste'))
```
